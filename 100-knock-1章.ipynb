#02. 「パトカー」＋「タクシー」＝「パタトクカシーー」

#####「パトカー」＋「タクシー」の文字を先頭から交互に連結して文字列「パタトクカシーー」を得よ．

```python
str1 = "パトカー"
str2 = "タクシー"
new_str = ""
for p, t in zip(str1, str2):
  new_str += p + t

print(new_str)
```

#03. 円周率

#####“Now I need a drink, alcoholic of course, after the heavy lectures involving quantum mechanics.”という文を単語に分解し，各単語の（アルファベットの）文字数を先頭から出現順に並べたリストを作成せよ．

```python
sentence = "Now I need a drink, alcoholic of course, after the heavy lectures involving quantum mechanics."
words = sentence.replace(',', ' ').replace('.', ' ').split()

word_length_list = list()

for word in words:
  word_length_list.append(len(word))

print(word_length_list)
```

#04. 元素記号

#####“Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can.”という文を単語に分解し，1, 5, 6, 7, 8, 9, 15, 16, 19番目の単語は先頭の1文字，それ以外の単語は先頭の2文字を取り出し，取り出した文字列から単語の位置（先頭から何番目の単語か）への連想配列（辞書型もしくはマップ型）を作成せよ．

```python
sentence = "Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can."
words = sentence.split()
dict = {}

for i, word in enumerate(words, 1):
  if i in [1, 5, 6, 7, 8, 9, 15, 16, 19]:
    dict[i] = word[:1]
  else:
    dict[i] = word[:2]

print(dict)
```

#05. n-gram

#####与えられたシーケンス（文字列やリストなど）からn-gramを作る関数を作成せよ．この関数を用い，”I am an NLPer”という文から単語bi-gram，文字bi-gramを得よ．

```python
def n_gram(target, n):
  # 基準を1文字(単語)ずつ ずらしながらn文字分抜き出す
  return [ target[idx:idx + n] for idx in range(len(target)-1)]

target = "I am an NLPer"
print(n_gram(target, 2))


words = target.split(' ')
print(n_gram(words, 2))
```

#06. 集合

#####“paraparaparadise”と”paragraph”に含まれる文字bi-gramの集合を，それぞれ, XとYとして求め，XとYの和集合，積集合，差集合を求めよ．さらに，’se’というbi-gramがXおよびYに含まれるかどうかを調べよ．

```python
def bi_gram(text):
    return {text[i:i+2] for i in range(len(text)-1)}

text_x = "paraparaparadise"
text_y = "paragraph"

X = bi_gram(text_x)
Y = bi_gram(text_y)

union_set = X | Y
intersection_set = X & Y
difference_set = X - Y

se_in_X = 'se' in X
se_in_Y = 'se' in Y

print(f"X: {X}")
print(f"Y: {Y}")
print(f"XとYの和集合: {union_set}")
print(f"XとYの積集合: {intersection_set}")
print(f"XとYの差集合(X - Y): {difference_set}")
print(f"’se’がXに含まれるか: {se_in_X}")
print(f"’se’がYに含まれるか: {se_in_Y}")
```

#08. 暗号文

#####与えられた文字列の各文字を，以下の仕様で変換する関数cipherを実装せよ．

- 英小文字ならば(219 - 文字コード)の文字に置換
- その他の文字はそのまま出力

#####この関数を用い，英語のメッセージを暗号化・復号化せよ．

```python
def cipher(text):
    result = ""
    for char in text:
        if char.islower():
            result += chr(219 - ord(char))
        else:
            result += char
    return result

# テストデータ
original_message = "Hello, World!"

# 暗号化
encryption_message = cipher(original_message)
print("暗号化されたメッセージ: " + encryption_message)

# 復号化
decryption_message = cipher(encryption_message)
print("復号化されたメッセージ: " + decryption_message)
```

#09. Typoglycemia

#####スペースで区切られた単語列に対して，各単語の先頭と末尾の文字は残し，それ以外の文字の順序をランダムに並び替えるプログラムを作成せよ．ただし，長さが４以下の単語は並び替えないこととする．適当な英語の文（例えば”I couldn’t believe that I could actually understand what I was reading : the phenomenal power of the human mind .”）を与え，その実行結果を確認せよ．

```python
import random

def random_word(word):
    if len(word) <= 4:
        return word
    else:
        middle_part = list(word[1:-1])
        random.shuffle(middle_part)
        return word[0] + ''.join(middle_part) + word[-1]

sentence = "I couldn’t believe that I could actually understand what I was reading : the phenomenal power of the human mind."
words = sentence.split()
random_words = [random_word(word) for word in words]
random_sentence = ' '.join(random_words)

print("並び替えた文:", random_sentence)
```
